```python
import pandas as pd
import numpy as np
from sklearn.model_selection import cross_val_predict, cross_val_score,KFold
from sklearn.model_selection import cross_val_predict
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
from sklearn.metrics import make_scorer, accuracy_score, precision_score, recall_score, f1_score, confusion_matrix
from sklearn.neighbors import KNeighborsClassifier
from sklearn.model_selection import StratifiedKFold
from sklearn.model_selection import cross_validate
import matplotlib.pyplot as plt
```

```python

path ="/content/drive/MyDrive/CPA1.csv"

df = pd.read_csv(path)

data_array = df.values

print(data_array)


```

```python
X = data_array[:, :-1]  # Ambil semua kolom kecuali kolom terakhir
y = data_array[:, -1]   # Ambil kolom terakhir sebagai label
print("Fitur (X):\n", X)
print("Label (y):\n", y)
```

```python
#  Bagi data menjadi set pelatihan dan pengujian (75% pelatihan, 25% pengujian)SVM
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=42)

# 3. Latih model SVM
svm_model = SVC(kernel='linear')  # Anda dapat memilih kernel yang sesuai
svm_model.fit(X_train, y_train)

# 4. Evaluasi model dan hitung metrik performansi
y_pred = svm_model.predict(X_test)

# Hitung confusion matrix
cm = confusion_matrix(y_test, y_pred)

# Hitung Sensitivity, Specificity, Accuracy, dan F1-Score
TP = cm[1, 1]
TN = cm[0, 0]
FP = cm[0, 1]
FN = cm[1, 0]

sensitivity = TP / (TP + FN)
specificity = TN / (TN + FP)
accuracy = accuracy_score(y_test, y_pred)
precision = TP/(TP+FP)
f1=2*precision*sensitivity/(precision+sensitivity)


print("Sensitivity:", sensitivity)
print("Specificity:", specificity)
print("Accuracy:", accuracy)
print("precision:", precision)
print("F1-Score:", f1)

metrics = ['accuracy', 'precision', ' Sensitivity', 'Specificity', 'F1-Score']
values = [accuracy, precision, sensitivity, specificity, f1]
# Plot bar chart
plt.figure(figsize=(2, 1))
plt.bar(metrics, values, color=['blue', 'green', 'orange', 'red', 'purple'])
plt.title(' SVM 75 train, 25 test ')
plt.xlabel('Metrics')
plt.ylabel('Values')
plt.ylim(0, 1)  # Memastikan skala sumbu y dari 0 hingga 1
plt.xticks(rotation=45)  # Memutar label sumbu x untuk memastikan kejelasan
plt.show()
```
